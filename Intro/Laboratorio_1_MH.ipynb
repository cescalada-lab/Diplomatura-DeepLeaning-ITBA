# Monty Hall

```python
import numpy as np
from matplotlib import pyplot as plt

#Creamos las varibles para iniciar el juego
Ganar_Cambiando=0
Ganar_NoCambiando=0
Ganar_Aleatoriamente=0

#Variables que acumulan las cantidad de veces que se ganó
Acum_Ganar_Cambiando=[]
Acum_Ganar_NoCambiando=[]
Acum_Ganar_Aleatoriamente=[]

#Cantidad de Iteraciones
N=1000

#Iniciamos el juego
for i in range(N):
    Puerta = [1,2,3]                                                     #Cantidad de puertas
    Puerta_Auto= np.random.choice(Puerta)                                #Elegimos la ubicación del auto aleatoriamente
    Puerta_Usuario= np.random.choice(Puerta)                             #El usuario elige una puerta
    
    #El host elige una puerta:
    #Caso que no coincidan la puerta del auto con la que eligió el usuario
    if Puerta_Auto != Puerta_Usuario:
        (Puerta.remove(Puerta_Auto) , Puerta.remove(Puerta_Usuario))     #Se eliminan ambas para ver cual queda
        Puerta_Eliminar= np.random.choice(Puerta)                        #Se selecciona la única que queda 
        Puerta.remove(Puerta_Eliminar)                                   #Host abre la puerta
        #Usuario decide si quiere cambiar/mantener:   
        #Usuario si cambia siempre:
        Ganar_Cambiando=Ganar_Cambiando+1
        #Usuario elije Cambiar/Mantener Aleatoriamente:
        Seleccion_Aleatoria=np.random.randint(0,2)                       #Convención: 0=decide cambiar de puerta,
        if Seleccion_Aleatoria == 0:                                     #            1=decide mantener puerta
            Ganar_Aleatoriamente=Ganar_Aleatoriamente+1
        
    #Caso que si coincidan la puerta del auto con la que eligió el usuario    
    if Puerta_Auto == Puerta_Usuario:
        ( Puerta.remove(Puerta_Auto) )                                   #Esta Puerta se elimina porque ya fue elegida
        Puerta_Eliminar= np.random.choice(Puerta)                        #Se elige una puerta al azar
        Puerta.remove(Puerta_Eliminar)                                   #El host abre la puerta que eligió
        #Usuario decide si quiere cambiar/mantener:
        #Usuario no cambia nunca:
        Ganar_NoCambiando=Ganar_NoCambiando+1
        #Usuario elije Cambiar/Mantener Aleatoriamente:
        Seleccion_Aleatoria=np.random.randint(0,2)                       #Convención: 0=decide cambiar de puerta, 
        if Seleccion_Aleatoria == 1:                                     #            1=decide mantener puerta
            Ganar_Aleatoriamente=Ganar_Aleatoriamente+1
    
    Acum_Ganar_Cambiando.append(Ganar_Cambiando)
    Acum_Ganar_NoCambiando.append(Ganar_NoCambiando)
    Acum_Ganar_Aleatoriamente.append(Ganar_Aleatoriamente)
  
    
#Muestro Resulados
print("Ganar Cambiando:",Ganar_Cambiando)
print("Ganar Manteniendo:",Ganar_NoCambiando)
print("Ganar Aleatoriamente:",Ganar_Aleatoriamente)
print("\n")

#Hago el Gráfico
plt.figure(figsize=(8,4))
plt.plot(Acum_Ganar_Cambiando, label="Ganar Cambiando")
plt.plot(Acum_Ganar_NoCambiando, label="Ganar Manteniendo")
plt.plot(Acum_Ganar_Aleatoriamente, label="Ganar Aleatoriamente")
plt.title("Monty Hall Gráfico", fontsize=16,fontweight="bold")
plt.xlabel("Cantidad de Iteraciones (X)")
plt.ylabel("Aciertos (Y)")
plt.legend()
plt.grid()
plt.show()
```
